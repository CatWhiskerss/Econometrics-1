```python
b=np.random.randint(1,4)
data2 = data.loc[data[11] == b].dropna().reset_index(drop = True)
data2
```

Работаем с двумя фиктивными переменными

```python
a=np.random.randint(0,2)
y=data2.iloc[:,a].values
x=data2.iloc[:,2:11].values
x = StandardScaler().fit_transform(x)
pca = PCA(n_components = 2)
XPCAreduced = pca.fit_transform(x)
XPCAreduced
```

Строим линейную регрессию

```python
model = lm.LinearRegression()
model.fit(XPCAreduced, y)
model.intercept_, model.coef_
```
